```python
import pandas as pd
```

```python
PRESIDENT_DATA_PATH = ('1976-2020-president.csv')
data_president_df = pd.read_csv(PRESIDENT_DATA_PATH)
data_president_df
```

```python
data_president_year_2020_df = data_president_df[data_president_df['year'] == 2020]
data_president_year_2020_df
```

```python
columns_relevant = ['year', 'state', 'candidate', 'candidatevotes', 'party_simplified']
data_president_columns_filtered_df = data_president_year_2020_df[columns_relevant]
data_president_columns_filtered_df
```

```python
state_party_sum_of_votes = data_president_columns_filtered_df.groupby(['state', 'party_simplified'])['candidatevotes'].sum().reset_index()

index_party_bigger_vote_by_state = state_party_sum_of_votes.groupby(['state'])['candidatevotes'].transform(max) == state_party_sum_of_votes['candidatevotes']
most_votes_by_party = state_party_sum_of_votes[index_party_bigger_vote_by_state]

# Se houver empates, isso manterÃ¡ apenas a primeira ocorrÃªncia
most_votes_by_party = most_votes_by_party.drop_duplicates(subset=['state'], keep='first')
most_votes_by_party
```
